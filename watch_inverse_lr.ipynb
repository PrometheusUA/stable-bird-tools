```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_lr(epoch, warmup=0.99, inv_gamma=1000, power=0.5, final_lr=0., base_lr=1e-4):
    warmup = 1 - warmup ** (epoch + 1)
    lr_mult = (1 + epoch / inv_gamma) ** -power
    return warmup * max(final_lr, base_lr * lr_mult)
```

```python
epochs = np.arange(0, 10000)
```

```python
lrs = [get_lr(epoch) for epoch in epochs]
```

```python
plt.plot(epochs, lrs)
```
